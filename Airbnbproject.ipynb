```python
import numpy as np
import pandas as pd
import datetime

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
x = data.iloc[:,[5]]
```

```python
x
```

```python
y = data.iloc[:,[1]]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
x_pred = reg.predict(x_train)
```

```python
plt.scatter(x_train,y_train, color='yellow')
plt.plot(x_train,x_pred,color='red')
plt.show()
print("training set")
```

```python
plt.scatter(x_test,y_test, color='blue')
plt.plot(x_train,x_pred,color='red')
plt.show()
print("test set")
```

```python
import sklearn.metrics as sm
print("Mean absolute error=", round(sm.mean_absolute_error(y_test,y_pred),2))
print("Mean squared error =",round(sm.mean_squared_error(y_test,y_pred),2))
print("Median absolute error =", round(sm.median_absolute_error(y_test,y_pred), 2))
print("Explain Variance Score =", round(sm.explained_variance_score(y_test,y_pred), 2))
print("R2 Score =", round(sm.r2_score(y_test,y_pred),2))
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st_x=StandardScaler()
x1_train=st_x.fit_transform(x1_train)
x1_test=st_x.transform(x1_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(x1_train,y1_train)
pred1=knn.predict(x1_test)
print(pred1)
```

```python
%matplotlib inline
import seaborn as sns
```

```python

ax1=sns.distplot(y1_test,hist=False,color='r',label='actual value')
ax2=sns.distplot(pred1,hist=False,color='blue',label='predicted value')
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import neighbors
from math import sqrt
rmse=list()
rmse1=list()
new=list()
mylist=list(range(50))
mylist=list(filter(lambda x:x%2!=0,mylist))
for k in mylist:
    knn=KNeighborsRegressor(n_neighbors=k)
    knn.fit(x1_train,y1_train)
    pred1=knn.predict(x1_test)
    pred2=knn.predict(x1_train)
    error=sqrt(mean_squared_error(y1_test,pred1))
    error1=sqrt(mean_squared_error(y1_train,pred2))
    rmse.append(error)
    new.append([error,k])
    rmse1.append(error1)
    print('RMSE test value for k= ' , k , 'is:', error, '\nRMSE train value for k is = ',error1)
```

```python
newList=list()
for x in mylist:
    newList.append(1/x)
```

```python
Rmsetest=min(new)
print("Minimum R squared test error=",Rmsetest[0])
Rmsetrain=min(rmse1)
print("Minimum R squared train error=",Rmsetrain)
print("Optimal K value= ",Rmsetest[1])
```
